# O que é One Hot Encoding?

>One Hot Encoding é uma técnica para transformar dados categóricos (palavras, rótulos, nomes) em números binários, de modo que os algoritmos de machine learning possam entender e processar esses dados.

```python
dados = ['gato', 'cachorro', 'pássaro', 'rato', 'foca', 'peixe', 'cobra', 'formiga']
```

```python
encoding = {}

# Passo 1: Percorrer as categorias e seus respectivos índices
for idx, categoria in enumerate(dados):
    vetor = []  # Inicia um vetor vazio para esta categoria

    # Passo 2: Criar o vetor one-hot para a categoria atual
    for i in range(len(dados)):
        if i == idx:
            vetor.append(1)
            print(categoria)
            print(vetor)
        else:
            vetor.append(0)


    # Passo 3: Adicionar ao dicionário de encoding
    encoding[categoria] = vetor
```

# Exemplo de uso:

```python
def one_hot_encode(data):
    # Identifica categorias únicas
    categorias = list(set(data))
    categorias.sort()  # Para manter ordem previsível

    # Mapeia cada categoria para um vetor
    encoding = {categoria: [1 if i == idx else 0 for i in range(len(categorias))]
                for idx, categoria in enumerate(categorias)}


    return encoding
```

```python
mapeamento = one_hot_encode(dados)
```

```python
print("Mapeamento de categorias:")
for chave, valor in mapeamento.items():
    print(f"{chave}: {valor}")
```
